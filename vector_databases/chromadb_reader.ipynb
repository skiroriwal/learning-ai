```python
import chromadb

### created chromadb client
chromadb_client = chromadb.PersistentClient(path="./disk_chromadb")

### created 3 collections
collection1 = chromadb_client.get_collection(name="collection1")

### show collections
collections = chromadb_client.list_collections()
print(collections)
```

```python
results = collection1.query(
    query_texts=["what is Sachin Tendulkar record in world cup"],
    n_results=2,
)
print(results)
```
